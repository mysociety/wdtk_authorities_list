## Example table

This shows a table being rendered as markdown

```python
source = pd.DataFrame(
    {
        "Vertical": ["A", "B", "C", "D", "E", "F", "G", "H", "I"],
        "Horizontal": [28, 55, 43, 91, 81, 53, 19, 87, 52],
    }
)

source["double_h"] = source["Horizontal"] * 2

source
```

# Example chart

```python
chart = (
    Chart(source)
    .mark_bar()
    .encode(y="Vertical", x="Horizontal")
    .properties(
        title="This is an example chart, with the right fonts",
        logo=True,
        caption="Data source goes here",
        width=800,
    )
)

chart.display()
```
